```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from datetime import datetime
import locale
locale.setlocale(locale.LC_ALL, 'pt_BR.utf8')  # Set locale to brasil
# Start session
driver = webdriver.Firefox()
```

```python
# Get to the g1 site
driver.get("https://g1.globo.com/")
driver.title
```

```python
# Search for Copa 2010
search_box = driver.find_element(by=By.ID, value="busca-campo")
search_box.send_keys("Copa 2010")
search_box.submit()
```

```python
# Activate the range data
filter_data = driver.find_element(by=By.CLASS_NAME, value="filters__advanced-date-filter")
filter_data.click()
range_data = driver.find_element(by=By.CLASS_NAME, value="filters__dropdown__list__range-date")
range_data.click()

# Define range data
start = datetime.strptime("14/04/2019", "%d/%m/%Y")
end = datetime.strptime("18/05/2022", "%d/%m/%Y")
```

```python
def travel_to_month(expected_date:datetime):
    month = driver.find_element(by=By.XPATH, value="//div[@class='CalendarMonth CalendarMonth_1'][ @data-visible='true']/div[@class='CalendarMonth_caption CalendarMonth_caption_1']")
    raw_month = month.find_element(by=By.TAG_NAME, value="strong").get_attribute('innerHTML')
    # Convert to date
    actual_month = datetime.strptime(raw_month, "%B %Y")
    while actual_month.year != expected_date.year or \
          actual_month.month != expected_date.month:
        # print(actual_month)
        if (actual_month < expected_date):
            # Move forward
            xpath = "/html/body/div[2]/div/div[1]/div/div/div[2]/div[1]/div/div[2]/div/div/div/div[2]/div[1]/div[2]"
        else:
            # Move Backward
            xpath = "/html/body/div[2]/div/div[1]/div/div/div[2]/div[1]/div/div[2]/div/div/div/div[2]/div[1]/div[1]"
        change_month = driver.find_element(by=By.XPATH, value=xpath)
        change_month.click()
        month = driver.find_element(by=By.XPATH, value="//div[@class='CalendarMonth CalendarMonth_1'][ @data-visible='true']/div[@class='CalendarMonth_caption CalendarMonth_caption_1']")
        raw_month = month.find_element(by=By.TAG_NAME, value="strong").get_attribute('innerHTML')
        # Convert to date
        actual_month = datetime.strptime(raw_month, "%B %Y")

def select_day(date:datetime):
    str_day = date.strftime("%d de %B de %Y")
    table_month = driver.find_element(by=By.XPATH, value="//div[@class='CalendarMonth CalendarMonth_1'][ @data-visible='true']/table")
    table_day = table_month.find_element(by=By.XPATH, value=f"//td[@aria-label[contains(., '{str_day}')]][text()='{date.day}']")
    ActionChains(driver).move_to_element(table_day).click(table_day).perform()
```

```python
travel_to_month(start)
select_day(start)
travel_to_month(end)
select_day(end)
```

```python
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from datetime import datetime
import locale
from time import sleep
locale.setlocale(locale.LC_ALL, 'pt_BR.utf8')  # Set locale to brasil
# Start session
options = webdriver.FirefoxOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('headless')
driver = webdriver.Firefox(options=options)
```

```python
def scroll_shim(passed_in_driver, object):
    x = object.location['x']
    y = object.location['y']
    scroll_by_coord = 'window.scrollTo(%s,%s);' % (
        x,
        y
    )
    scroll_nav_out_of_way = 'window.scrollBy(0, -120);'
    passed_in_driver.execute_script(scroll_by_coord)
    passed_in_driver.execute_script(scroll_nav_out_of_way)
# Get to the g1 site
driver.get("https://veja.abril.com.br/2019/")
driver.title
```

```python
def readNews(i, file):
    title_elem = driver.find_element(by=By.CLASS_NAME, value="title")
    print(f"Reading[{i}]: {title_elem.text}")
    description = driver.find_element(by=By.CLASS_NAME, value="description")
    content_section = driver.find_element(by=By.CLASS_NAME, value="content")
    list_p = content_section.find_elements(by=By.TAG_NAME, value="p")
    list_p = list_p[:-3]

    file.write(title_elem.text + "\n")
    file.write(description.text + "\n")
    for p in list_p:
        file.write(p.text)
    file.write("\n\n")
    print()
```

```python
element_news_list = driver.find_element(by=By.ID, value='infinite-list')
news_list = element_news_list.find_elements(by=By.XPATH, value="//*[starts-with(@id, 'post')]")
```

```python
news = news_list[-1]
scroll_shim(driver, news)
ActionChains(driver).move_to_element(news).perform()
```

```python
from time import sleep
i = 0
while True:
    element_news_list = driver.find_element(by=By.ID, value='infinite-list')
    news_list = element_news_list.find_elements(by=By.XPATH, value="//*[starts-with(@id, 'post')]")
    news = news_list[-1]
    scroll_shim(driver, news)
    ActionChains(driver).move_to_element(news).perform()

    try:
        more = driver.find_element(by=By.ID, value="infinite-handle")
        btn = more.find_element(by=By.TAG_NAME, value="button")
        if btn.is_displayed():
            btn.click()
        else:
            print("Something went wrong")
            break
    except:
        print("Done")
        break
    sleep(0.5)
```

# Pega todos os links

```python
element_news_list = driver.find_element(by=By.ID, value='infinite-list')
news_list = element_news_list.find_elements(by=By.XPATH, value="//*[starts-with(@id, 'post')]")
with open("links-2020.txt", "w") as f:
    for n in news_list:
        link = n.find_element(by=By.TAG_NAME, value="a").get_attribute("href")
        f.write(link + "\n")
```

```python
with open("links-2020.txt", "r") as f, \
         open("Veja-2020.txt", "a+") as w:
    for i, link in enumerate(f.readlines()):
        if i >= 905:
            print(f"Link: {link[:-1]}")
            driver.get(link)
            readNews(i, w)
            sleep(1)
```

```python
SCROLL_PAUSE_TIME = 3

from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException

# Get scroll height
last_height = driver.execute_script("return document.body.scrollHeight")
h = 100

while True:
    # Scroll down to bottom
    driver.execute_script(f"window.scrollTo(0, {h});")

    # Wait to load page
    sleep(SCROLL_PAUSE_TIME)
    try:
        more = driver.find_element(by=By.ID, value="infinite-handle")
        ActionChains(driver).move_to_element(more).click(more).perform()
    except NoSuchElementException:
        pass
    # Calculate new scroll height and compare with last scroll height
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    h += 100
```
